```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

plt.rcParams.update({'font.size': 22})
```

```python
dp_cold = ["../data/onoff_1.raw",
                 "../data/onoff_3.raw",
                 "../data/onoff_4.8.raw",
                 "../data/onoff_8.raw"]

# set up the figure
plt.figure(figsize=(10,8))
plt.title("On-off control for -5$^{\circ}$C")

data = np.genfromtxt(dp_cold[0], skip_header=13, delimiter=",")
# select the starting point where the error of the temp is close to zero
start = int(np.where((data[:,2] > -5e-2) & (data[:,2] < 5e-2))[0][3])
# choose the endpoint so that all files have the same length
end = start + 650
plt.plot((data[start:end,0]-data[start,0]), data[start:end,1], ".",label = "Band 1$^{\circ}$C")


data = np.genfromtxt(dp_cold[1], skip_header=13, delimiter=",")
start = int(np.where((data[:,2] > -5e-2) & (data[:,2] < 5e-2))[0][0])
end = start + 650
plt.plot((data[start:end,0]-data[start,0]), data[start:end,1], ".", label = "Band 3$^{\circ}$C")


data = np.genfromtxt(dp_cold[2], skip_header=13, delimiter=",")
start = int(np.where((data[:,2] > -5e-2) & (data[:,2] < 5e-2))[0][3])
end = start + 650
plt.plot((data[start:end,0]-data[start,0]), data[start:end,1], ".", label = "Band 4.8$^{\circ}$C")

data = np.genfromtxt(dp_cold[3], skip_header=13, delimiter=",")
start = int(np.where((data[:,2] > -5e-2) & (data[:,2] < 5e-2))[0][3])
end = start + 650
plt.plot((data[start:end,0]-data[start,0]), data[start:end,1], ".", label = "Band 8$^{\circ}$C")

plt.legend()
plt.show()
```

```python
dp_cold = ["../data/onoff_1.raw",
                 "../data/onoff_3.raw",
                 "../data/onoff_4.8.raw",
                 "../data/onoff_8.raw"]

# set up the figure
f, (ax1, ax2, ax3, ax4) = plt.subplots(4,1, sharex=True, figsize=(14,8))
plt.subplots_adjust(left=None, bottom=0.1, right=None, top=0.88, wspace=None, hspace=0.5)
plt.suptitle("On-off control for -5$^{\circ}$C", fontsize="x-large")

ax1.set_title("1$^{\circ}$C band", fontsize="small")
ax2.set_title("3$^{\circ}$C band", fontsize="small")
ax3.set_title("4.8$^{\circ}$C band", fontsize="small")
ax4.set_title("8$^{\circ}$C band", fontsize="small")

ax4.set_xlabel("Time [s]")
f.text(0.04, 0.5, 'Temperature [$^{\circ}$C]', va='center', rotation='vertical')

# First subplot: band 1
data = np.genfromtxt(dp_cold[0], skip_header=13, delimiter=",")
# select the starting point where the error of the temp is close to zero
start = int(np.where((data[:,2] > -5e-2) & (data[:,2] < 5e-2))[0][3])
# choose the endpoint so that all files have the same length
end = start + 500
ax1.plot((data[start:end,0]-data[start,0]), data[start:end,1], "C0-",label = "Band 1$^{\circ}$C")

onecycle = np.where((data[:,2] > -5e-2) & (data[:,2] < 5e-2))[0][8]
ax1.plot((data[onecycle,0]-data[start,0]), data[onecycle,1], "C0o")

# 2nd subplot: band 3
data = np.genfromtxt(dp_cold[1], skip_header=13, delimiter=",")
start = int(np.where((data[:,2] > -5e-2) & (data[:,2] < 5e-2))[0][0])
end = start + 500
ax2.plot((data[start:end,0]-data[start,0]), data[start:end,1], "C1-.", label = "Band 3$^{\circ}$C")
onecycle = np.where((data[:,2] > -5e-2) & (data[:,2] < 5e-2))[0][4]
ax2.plot((data[onecycle,0]-data[start,0]), data[onecycle,1], "C1o")

# 3rd subplot: band 4.8
data = np.genfromtxt(dp_cold[2], skip_header=13, delimiter=",")
start = int(np.where((data[:,2] > -5e-2) & (data[:,2] < 5e-2))[0][3])
end = start + 500
ax3.plot((data[start:end,0]-data[start,0]), data[start:end,1], "C2-", label = "Band 4.8$^{\circ}$C")
onecycle = np.where((data[:,2] > -5e-2) & (data[:,2] < 5e-2))[0][8]
ax3.plot((data[onecycle,0]-data[start,0]), data[onecycle,1], "C2o")

# 4th subplot: band 8
data = np.genfromtxt(dp_cold[3], skip_header=13, delimiter=",")
start = int(np.where((data[:,2] > -5e-2) & (data[:,2] < 5e-2))[0][3])
end = start + 500
ax4.plot((data[start:end,0]-data[start,0]), data[start:end,1], "C3--", label = "Band 8$^{\circ}$C")
onecycle = np.where((data[:,2] > -5e-2) & (data[:,2] < 5e-2))[0][7]
ax4.plot((data[onecycle,0]-data[start,0]), data[onecycle,1], "C3o")

plt.show()
```
