```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

plt.rcParams.update({'font.size': 16})
```

# Minimal temp

```python
path = "../data/openloop_mintemp4095.raw"
plt.figure(figsize=(12,3))
plt.title("Open-loop operation: minimum temperature")
data = np.genfromtxt(path, skip_header=13, delimiter=",")
plt.ylabel("Temperature [$^{\circ}$C]")
plt.xlabel("Time [ms]")
plt.plot(data[:,0]-data[0,0], data[:,1])
plt.show()
```

```python
print(np.mean(data[:,1][-20:]))
print(np.std(data[:,1][-20:]))
```

# Maximal temp

```python
path = "../data/openloop_maxtemp4095.raw"
plt.figure(figsize=(12,3))
plt.title("Open-loop operation: maximum temperature")
data = np.genfromtxt(path, skip_header=13, delimiter=",")
plt.ylabel("Temperature [$^{\circ}$C]")
plt.xlabel("Time [ms]")
plt.plot(data[:-45,0]-data[0,0], data[:-45,1])
plt.show()
```

```python
print(np.mean(data[:,1][-10:]))
print(np.std(data[:,1][-10:]))
```

# -5C

```python
path = "../data/openloop_-5_2275.raw"
plt.figure(figsize=(12,3))
plt.title("Open-loop operation: reaching -5$^{\circ}$C")
data = np.genfromtxt(path, skip_header=13, delimiter=",")
plt.ylabel("Temperature [$^{\circ}$C]")
plt.xlabel("Time [ms]")
plt.plot(data[2000:-45,0]-data[0,0], data[2000:-45,1])
plt.show()
```

```python
print(np.mean(data[:,1][-10:]))
print(np.std(data[:,1][-10:]))
```
